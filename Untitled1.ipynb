```python

!pip install pybullet stable-baselines3 gymnasium numpy matplotlib torch


class Config:
    MAX_TIMESTEPS = 1000
    GOAL_THRESHOLD = 1.5
    COLLISION_PENALTY = -100
    GOAL_REWARD = 200
    STEP_PENALTY = -0.5
    LIDAR_RAYS = 16
    LIDAR_MAX_DIST = 15.0
    DRONE_MASS = 1.0
    MAX_VELOCITY = 3.0
    WORLD_SIZE = 20
    NUM_OBSTACLES = 8
    OBSTACLE_MIN_RADIUS = 0.5
    OBSTACLE_MAX_RADIUS = 2.0
    TOTAL_TIMESTEPS = 500_000
    LEARNING_RATE = 3e-4
    N_STEPS = 2048
    BATCH_SIZE = 64
    N_EPOCHS = 10
    GAMMA = 0.99
    GAE_LAMBDA = 0.95
    CLIP_RANGE = 0.2
    ENT_COEF = 0.01
    EVAL_EPISODES = 10
    RENDER_EVAL = True
    MODEL_SAVE_PATH = "models/drone_ppo_model"
    LOG_DIR = "logs/"

# CELL 3: Environment (NumPy Physics)
import gymnasium as gym
import numpy as np
from gymnasium import spaces
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

class DroneNavEnv(gym.Env):
    metadata = {"render_modes": ["human", "rgb_array"], "render_fps": 30}

    def __init__(self, render_mode=None):
        super().__init__()
        self.render_mode = render_mode
        self.cfg = Config()
        self.action_space = spaces.Box(low=-1.0, high=1.0, shape=(3,), dtype=np.float32)
        obs_dim = self.cfg.LIDAR_RAYS + 3 + 3
        self.observation_space = spaces.Box(low=-10.0, high=10.0, shape=(obs_dim,), dtype=np.float32)
        self.drone_pos = None
        self.drone_vel = None
        self.goal_pos = None
        self.obstacles = []
        self.current_step = 0
        self.prev_dist_to_goal = None
        self.path_history = []
        self.fig = None
        self.ax = None

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        self.path_history = []
        self.goal_pos = np.array([
            np.random.uniform(5, self.cfg.WORLD_SIZE/2),
            np.random.uniform(5, self.cfg.WORLD_SIZE/2),
            np.random.uniform(2, 8)
        ], dtype=np.float32)
        self.obstacles = []
        attempts = 0
        while len(self.obstacles) < self.cfg.NUM_OBSTACLES and attempts < 100:
            attempts += 1
            pos = np.random.uniform(-self.cfg.WORLD_SIZE/2, self.cfg.WORLD_SIZE/2, size=3)
            pos[2] = np.random.uniform(1.0, 8.0)
            radius = np.random.uniform(self.cfg.OBSTACLE_MIN_RADIUS, self.cfg.OBSTACLE_MAX_RADIUS)
            if np.linalg.norm(pos[:2]) < 3.0: continue
            if np.linalg.norm(pos - self.goal_pos) < 3.0: continue
            too_close = False
            for ep, er in self.obstacles:
                if np.linalg.norm(pos - ep) < (radius + er + 1.0):
                    too_close = True
                    break
            if too_close: continue
            self.obstacles.append((pos.astype(np.float32), radius))
        self.drone_pos = np.array([0.0, 0.0, 2.0], dtype=np.float32)
        self.drone_vel = np.array([0.0, 0.0, 0.0], dtype=np.float32)
        self.prev_dist_to_goal = np.linalg.norm(self.drone_pos - self.goal_pos)
        obs = self._get_observation()
        info = {"goal_position": self.goal_pos.copy(), "num_obstacles": len(self.obstacles)}
        return obs, info

    def _get_observation(self):
        lidar = self._ray_casting()
        goal_vec = self.goal_pos - self.drone_pos
        dist = np.linalg.norm(goal_vec)
        if dist > 0.01: goal_vec = goal_vec / dist
        else: goal_vec = np.zeros(3, dtype=np.float32)
        vel = self.drone_vel.copy()
        return np.concatenate([lidar, goal_vec, vel]).astype(np.float32)

    def _ray_casting(self):
        num_rays = self.cfg.LIDAR_RAYS
        max_dist = self.cfg.LIDAR_MAX_DIST
        angles = np.linspace(0, 2*np.pi, num_rays, endpoint=False)
        distances = np.ones(num_rays, dtype=np.float32) * max_dist
        for i, angle in enumerate(angles):
            z_offset = np.sin(angle * 2) * 0.3
            direction = np.array([np.cos(angle), np.sin(angle), z_offset], dtype=np.float32)
            direction = direction / (np.linalg.norm(direction) + 1e-8)
            for obs_pos, obs_radius in self.obstacles:
                L = obs_pos - self.drone_pos
                tca = np.dot(L, direction)
                if tca < 0: continue
                d2 = np.dot(L, L) - tca * tca
                if d2 > obs_radius ** 2: continue
                thc = np.sqrt(max(0, obs_radius ** 2 - d2))
                t0 = tca - thc
                if t0 > 0.01 and t0 < distances[i]:
                    distances[i] = t0
            if direction[2] < -0.001:
                t_ground = -self.drone_pos[2] / direction[2]
                if t_ground > 0.01 and t_ground < distances[i]:
                    distances[i] = t_ground
        return distances / max_dist

    def _check_collision(self):
        drone_radius = 0.3
        for obs_pos, obs_radius in self.obstacles:
            if np.linalg.norm(self.drone_pos - obs_pos) < (obs_radius + drone_radius):
                return True
        if self.drone_pos[2] < 0.2 or self.drone_pos[2] > 15: return True
        if np.any(np.abs(self.drone_pos[:2]) > self.cfg.WORLD_SIZE): return True
        return False

    def _reached_goal(self):
        return np.linalg.norm(self.drone_pos - self.goal_pos) < self.cfg.GOAL_THRESHOLD

    def _compute_reward(self):
        dist_to_goal = np.linalg.norm(self.drone_pos - self.goal_pos)
        dist_improvement = self.prev_dist_to_goal - dist_to_goal
        reward = dist_improvement * 10.0 + self.cfg.STEP_PENALTY
        lidar = self._ray_casting()
        min_lidar = np.min(lidar)
        if min_lidar < 0.15: reward -= 20.0 * (0.15 - min_lidar)
        if self._check_collision(): reward += self.cfg.COLLISION_PENALTY
        if self._reached_goal(): reward += self.cfg.GOAL_REWARD
        reward -= 0.01 * abs(self.drone_vel[2])
        self.prev_dist_to_goal = dist_to_goal
        return reward

    def step(self, action):
        self.current_step += 1
        action = np.clip(action, -1.0, 1.0)
        target_acc = action * self.cfg.MAX_VELOCITY * 2.0
        dt = 0.05
        gravity = np.array([0.0, 0.0, -9.81], dtype=np.float32)
        self.drone_vel += (target_acc + gravity) * dt
        self.drone_vel = np.clip(self.drone_vel, -self.cfg.MAX_VELOCITY, self.cfg.MAX_VELOCITY)
        self.drone_vel *= 0.95
        self.drone_pos += self.drone_vel * dt
        self.path_history.append(self.drone_pos.copy())
        obs = self._get_observation()
        reward = self._compute_reward()
        terminated = self._check_collision() or self._reached_goal()
        truncated = self.current_step >= self.cfg.MAX_TIMESTEPS
        info = {"distance_to_goal": self.prev_dist_to_goal, "collision": self._check_collision(),
                "reached_goal": self._reached_goal(), "step": self.current_step}
        return obs, reward, terminated, truncated, info

    def close(self): pass

# CELL 4: Test
env = DroneNavEnv()
obs, info = env.reset()
print("Observation shape:", obs.shape)
print("Goal:", info['goal_position'])
for _ in range(10):
    action = env.action_space.sample()
    obs, reward, done, truncated, info = env.step(action)
    print(f"Step {info['step']}: Reward={reward:.2f}, Dist={info['distance_to_goal']:.2f}")
    if done: break
print("✅ Environment works!")

# CELL 5: Training
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import CheckpointCallback, EvalCallback
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.vec_env import DummyVecEnv, VecNormalize

def make_env():
    def _init():
        env = DroneNavEnv(render_mode=None)
        env = Monitor(env)
        return env
    return _init

env = DummyVecEnv([make_env()])
env = VecNormalize(env, norm_obs=True, norm_reward=True, clip_obs=10.0, clip_reward=10.0, gamma=Config.GAMMA)

model = PPO("MlpPolicy", env, learning_rate=Config.LEARNING_RATE, n_steps=Config.N_STEPS,
            batch_size=Config.BATCH_SIZE, n_epochs=Config.N_EPOCHS, gamma=Config.GAMMA,
            gae_lambda=Config.GAE_LAMBDA, clip_range=Config.CLIP_RANGE, ent_coef=Config.ENT_COEF,
            verbose=1, tensorboard_log=Config.LOG_DIR, device="auto")

print("🚀 Training started...")
model.learn(total_timesteps=100000)  # 100K for demo
model.save("drone_ppo_final")
env.save("vec_normalize.pkl")
print("✅ Training complete! Model saved.")

# CELL 6: Download Model
from google.colab import files
files.download('drone_ppo_final.zip')
files.download('vec_normalize.pkl')
```
